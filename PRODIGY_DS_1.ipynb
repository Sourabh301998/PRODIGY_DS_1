```
import zipfile
import pandas as pd

# Extract the zip
with zipfile.ZipFile("API_SP.POP.TOTL_DS2_en_csv_v2_81108.zip", 'r') as zip_ref:
    zip_ref.extractall("population_data")

# Load main CSV (skip top metadata rows)
df = pd.read_csv("population_data/API_SP.POP.TOTL_DS2_en_csv_v2_81108.csv", skiprows=4)

# Preview data
df.head()
```

```
import matplotlib.pyplot as plt

# Select data for 2023 and drop NaNs
pop_2023 = df[['Country Name', '2023']].dropna()

# Sort by population and get top 10
top10 = pop_2023.sort_values(by='2023', ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2023'], color='skyblue')
plt.title('Top 10 Most Populated Countries (2023)', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

To create a bar chart or histogram for age or gender distribution in a population, We first need a dataset that includes age or gender data. The World Bank file here uploaded contains total population, not individual age/gender breakdowns. So we used here sample gender distribution data.

```
import matplotlib.pyplot as plt

# Sample gender distribution data
gender_counts = {'Male': 480, 'Female': 520}

# Bar chart
plt.bar(gender_counts.keys(), gender_counts.values(), color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Sample age data (you can use real data instead)
ages = np.random.randint(0, 100, 1000)  # 1000 people with random ages between 0-99

# Histogram
plt.hist(ages, bins=10, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```
